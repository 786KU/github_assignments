# Line plot with multifacets

```python
# All plot are from the seaborn library gallery, you can copy the codes from there and use for your own datasets
# Link: https://seaborn.pydata.org/examples/index.html
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

nuqta=sns.load_dataset("dots")

nuqta.head()
```

```python
#defining a color palette
palette=sns.color_palette("rocket_r")

#plot line plot
sns.relplot(x="time", y="firing_rate",  
            hue="coherence", size="choice", col="align",
            kind="line", size_order=["T1", "T2"],
            palette=palette, height=5, aspect=0.75, 
            facet_kws=dict(sharex=False), data=nuqta)       # what is rely plot, check it
plt.show()
# this rely plot is like a graph changing with time, e.g., currency exchage plots on google, they are also interactive 
```

```python
# another type of box plot
import seaborn as sns
sns.set_theme(style="ticks", palette="pastel")

tips=sns.load_dataset("tips")
sns.boxplot(x="day", y="total_bill", hue="smoker", palette=["g", "m"], data=tips)
sns.despine(offset=10, trim=True) # despine is to cut the dots/outliers which are at the end of the data
plt.show()
```

```python
# another type of box plot
import seaborn as sns
sns.set_theme(style="ticks", palette="pastel")

tips=sns.load_dataset("tips")
sns.violinplot(x="day", y="total_bill", hue="smoker", 
               split=True, inner="quart", Linewidth=1, palette={"Yes":"b", "No":".85"}, data=tips)
sns.despine(left=True) # despine is to cut the dots/outliers which are at the end of the data
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme(style="whitegrid")

diamonds=sns.load_dataset("diamonds")
# draw a scatter plot whlie assigning point colors and sizes to different variables in the dataset

f,ax= plt.subplots(figsize=(6.5,6.5))
sns.despine(f, left=True, bottom=True)
clarity_ranking=["I1", "SI2", "SI1", "VS2", "VS1", "VVS2", "VVS1", "IF"]
sns.scatterplot(x="carat", y="price", hue="clarity", size="depth", 
                palette="ch:r=-.2,d=.3_r", hue_order=clarity_ranking, 
                sizes=(1,8), linewidth=0, data=diamonds, ax=ax)
plt.show()
```
