```python
kashti=sns.load_dataset("titanic")
sns.boxplot(x="survived", 
            y="age", 
           data=kashti,
            showmeans=True,
            meanprops={"marker":"*",
                       "markersize":"10", 
                       "markeredgecolor":"red",})
```

```python
# lets change the axis title etc

import matplotlib.pyplot as plt

kashti=sns.load_dataset("titanic")
sns.boxplot(x="survived", 
            y="age", 
           data=kashti,
            showmeans=True,
            meanprops={"marker":"*",
                       "markersize":"10", 
                       "markeredgecolor":"red",})

plt.xlabel("How many survived", size="10", weight="bold")
plt.ylabel("Age(years)", size="10", weight="bold")
plt.title("Titanic age vs Number of People survived", size="16", weight="bold")
```
