```python
# import libraries

import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# load dataset
phool= sns.load_dataset("iris")
phool
```

```python
# draw a line plot , line plot should be drawn between two numeric variables
sns.lineplot(x="sepal_length", y="sepal_width", data=phool)

# remember, we mostly plot numericall variables on yaxis and categorical variable on xaxis or in hue, we never take
# numerival varibale in the hue  
```

```python
# draw a line plot , line plot should be drawn between two numeric variables
sns.lineplot(x="sepal_length", y="sepal_width", data=phool)
plt.title("Phoolon ka plot")

# you can add title 
plt.title("Phoolon ka plot")

# You can limit the axis limits
plt.xlim(4)
plt.ylim(2)
```

```python
# you can change the style of line plot

# draw a line plot , line plot should be drawn between two numeric variables
sns.lineplot(x="sepal_length", y="sepal_width", data=phool)
plt.title("Phoolon ka plot")

# you can add title 
plt.title("Phoolon ka plot")

# You can limit the axis limits
plt.xlim(4)
plt.ylim(2)

sns.set_style("dark") # its not working, figure out why
```

```python
# how to set the style to show us dark background

# draw a line plot , line plot should be drawn between two numeric variables
sns.lineplot(x="sepal_length", y="sepal_width", data=phool)
plt.title("Phoolon ka plot")

# you can add title 
plt.title("Phoolon ka plot")

# You can limit the axis limits
plt.xlim(4)
plt.ylim(2)

# set style
sns.set_style("dark")
sns.set_style(style=None, rc=None) # now the background is dark
```
