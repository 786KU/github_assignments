```python
# how to remove the error bars?

# these bars are the confidence interval, and when you will set these intervals as none, these will be removed
import seaborn as sns
import matplotlib.pyplot as plt

# # lets load the titanic data to show grouping with barplot

kashti= sns.load_dataset("titanic")
kashti
# draw a bar plot,  you can change the order of x-axis man with woman etc.,you can also change the color palette 
sns.barplot(x="who", y="alone", hue="sex", data=kashti, order= ["child","woman", "man"], color="green", ci=None)
```

```python
# how to make horizontal bar plot
# you need to make the categorical axis on y-axis instead of xaxis, and numerial axis on the xaxis instead of yaxis
import seaborn as sns
import matplotlib.pyplot as plt

# # lets load the titanic data to show grouping with barplot

kashti= sns.load_dataset("titanic")
kashti
# draw a bar plot,  you can change the order of x-axis man with woman etc.,you can also change the color palette 
sns.barplot(x="alone", y="who", hue="sex", data=kashti, linewidth=3, facecolor=(1,.4,.6,1), edgecolor="0.5", errcolor="0.5") # you can search online which type of paleette you want to use
```
