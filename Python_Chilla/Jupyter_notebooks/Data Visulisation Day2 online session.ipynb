```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
```

```python

sns.set_theme(style="ticks", color_codes=True) # theme setting (style, color codes etc.), you can check the details on google
# for theme propoerties for syle and color codes etc
```

```python
titanic=sns.load_dataset("titanic") # laoding dataset from seaborn library, check on google about different datasets in seaborn
```

```python
sns.catplot(x="sex", y="survived", hue="class", kind="bar", data=titanic) # catplot means catogorical data plot
# rest is data(mapping, axis), geometries (kind), aysthetic (hue, color, shape, size etc.) 
```
